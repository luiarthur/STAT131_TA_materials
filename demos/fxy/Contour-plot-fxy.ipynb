```r
f = function(x, y) {
    # Indicator for whether (x, y) are in support
    insupport = (0 < y) & (y < 1 - x^2)
    
    # Evaluate function using indicator function notation.
    5/4 * (x^2 + y) * insupport
}
```

```r
# Creat grid points to evaluate (x, y)
x = seq(-1, 1, length= 300)
y = seq(0, 1, length= 300)

# Evaluate f at cartesian product of generated grid points (x, y)
z = outer(x, y, f)
```

```r
# Draw contour plots
contour(x, y, z, nlevels=30, ylim=0:1, xlab='x', ylab='y', main='Contour plot of f(x, y)',
        cex.main=1.5, cex.lab=1.5, cex.axis=1.5)
```

```r
# # Load a 3D plotting library
library(plot3D)
n = 10
x = seq(-1, 1, len=n)
y = seq(0, 1, len=n)
z = f(x, y)
M = mesh(x, y, z)
R = with(M, sqrt(x^2 + y^2 +z^2))

isosurf3D(x, y, z, colvar=R, level=0, col="red")
```
